```python
def binomialCoeff(n, k):
    # Create a 2D array to store results of subproblems
    C = [[0 for _ in range(k + 1)] for _ in range(n + 1)]

    # Calculate binomial coefficients in bottom-up manner
    for i in range(n + 1):
        for j in range(min(i, k) + 1):
            if j == 0 or j == i:
                C[i][j] = 1
            else:
                C[i][j] = C[i - 1][j - 1] + C[i - 1][j]

    return C[n][k]


# Main program
if __name__ == "__main__":
    n = int(input("Enter n: "))
    k = int(input("Enter k: "))
    print(f"Value of C({n}, {k}) is {binomialCoeff(n, k)}")
```
